## Creating categorical variables

In data science, a categorical variable is one that has a specific value from a limited selection of values. The number of values is usually fixed. Many developers will know categorical variables by the moniker enumerations. Each of the potential values that a categorical variable can assume is a level.



```python
import pandas as pd
car_colors = pd.Series(['Blue', 'Red', 'Green'], dtype='category')
car_data = pd.Series(pd.Categorical(['Yellow', 'Green', 'Red', 'Blue','Purple'], categories=car_colors, ordered=False))
find_entries = pd.isnull(car_data)
```

```python
print(car_colors)
print('---------------')
print(car_data)
print('---------------')
print(find_entries[find_entries == True])
```

**Renaming levels**

```python
car_colors = pd.Series(['Blue', 'Red', 'Green'], dtype='category')
car_data = pd.Series( 
    pd.Categorical(['Blue', 'Green', 'Red', 'Blue', 'Red'],
    categories=car_colors, ordered=False)
                    )

print(car_data)
print('-----------------After Renaming levels------------------')
car_colors.cat.categories = ["Purple", "Yellow", "Black"]
car_data.cat.categories = car_colors

print(car_data)
```

**Combining levels**

A particular categorical level might be too small to offer significant data for analysis. Perhaps there are only a few of the values, which may not be enough to create a statistical difference. In this case, combining several small categories might offer better analysis results.

**Concaneting**

```python

arr = np.array([4, 7, 12])
arr1 = np.array([5, 9, 15])

# Below are a quick examples

# Example 1: Use concatenate() to join two arrays
con = np.concatenate((arr, arr1))
print(con)
```

```python
arr = np.array([[4, 6],[9, 13]])
arr1 = np.array([[8, 3],[12, 19]])

# Example 2: Use concatenate() with axis
con = np.concatenate((arr, arr1), axis=1)
print(con)
```

```python
# Example 3: Use np.stack() function to Join Arrays
con = np.stack((arr, arr1), axis=0)
print(con)
```

```python
# Example 4: Use np.hstack() function
con = np.hstack((arr, arr1))
print(con)
```

```python
# Example 5: Use np.vstack() function 
con = np.vstack((arr, arr1))
print(con)
```

```python
# Example 6: Use np.dstack() function to Stacking Along Height (depth)
con = np.dstack((arr, arr1))
print(con)
```

## Variable - Cases

Officially, variable names in Python can be any length and can consist of uppercase and lowercase letters (A-Z, a-z), digits (0-9), and the underscore character (_). An additional restriction is that, although a variable name can contain digits, the first character of a variable name cannot be a digit.

**Camel Case:** Second and subsequent words are capitalized, to make word boundaries easier to see. 

**Example:** numberOfCollegeGraduates

**Pascal Case:** Identical to Camel Case, except the first word is also capitalized. 

**Example:** NumberOfCollegeGraduates

**Snake Case:** Words are separated by underscores.

**Example:** number_of_college_graduates

## Aggregating Data at Any Level

Aggregation is the process of combining or grouping data together into a
set, bag, or list. The data may or may not be alike. However, in most cases,
an aggregation function combines several rows together statistically using
­algorithms such as average, count, maximum, median, minimum, mode, or
sum.

There are several reasons to aggregate data:

1. Make it easier to analyze
2. Reduce the ability of anyone to deduce the data of an individual from the dataset for privacy or other reasons
3. Create a combined data element from one data source that matches a combined data element in another source

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'Map': [0,0,0,1,1,2,2], 'Values': [1,2,3,5,4,2,5]})
df['S'] = df.groupby('Map')['Values'].transform(np.sum)
df['M'] = df.groupby('Map')['Values'].transform(np.mean)
df['V'] = df.groupby('Map')['Values'].transform(np.var)

print(df)
```
